## Ptrecios de petroleo y monedas

```python
##Captura de las variables explicatorias únicamente
##Precios del petrolo de la OPEC
opec_petroleum_prices = quandl.get('OPEC/ORB')## Lamar la serie de tiempo
opec_petroleum_prices.rename(columns={'Value': 'Precio_petroleo_OPEC'}, inplace=True) ##Renombrar para cuando arme el dataset
opec_petroleum_prices.tail()## mostrar que se cuenta con datos actuales
```

```python
opec_petroleum_prices= opec_petroleum_prices.resample('M').mean() ##Cálculo del promedio mensual
opec_petroleum_prices.tail()
```

```python
## La oferta de monedas es un determinante de los precios de los bienes
eur_usd = quandl.get("FED/RXI_US_N_M_EU") ##Dolar USA
eur_usd.rename(columns={'Value': 'eur_usd'}, inplace=True)

yuan_usd = quandl.get("FED/RXI_N_M_CH") #Yuan China
yuan_usd.rename(columns={'Value': 'yuan_usd'}, inplace=True)
```

```python
####Mergear los datos de producción y oferta de crudo de los principales países
monedas = pd.merge(eur_usd, yuan_usd, on= 'Date')
monedas.tail()
```

```python
####Mergear los datos de producción y oferta de crudo de los principales países
precios = pd.merge(monedas, opec_petroleum_prices, on= 'Date')
precios.tail()
```

```python
##Exporto lo construido a Excel para facilitar ensayos
outputFile = "Data/precios.xlsx" ##Imprimiendo el conjunto de datos
precios.to_excel(outputFile) # Only argument is a string of the output file path
```

## Niveles de producción

```python
## Refinación de Fuel Oil en Canadá
refineryOutput_fuelOil_canada = quandl.get('JODI/OIL_RSROKB_CAN')
refineryOutput_fuelOil_canada.rename(columns={'Value': 'Refinación_fuelOil_canada'}, inplace=True)
refineryOutput_fuelOil_canada = refineryOutput_fuelOil_canada.drop(['Notes'], axis=1)
refineryOutput_fuelOil_canada.tail()
```

```python
### Producción de gas natural en Rusia
produccion_rusia_gas = quandl.get('JODI/OIL_NGPRKB_RUS')
produccion_rusia_gas.rename(columns={'Value': 'Produccion_gas_RUSIA'}, inplace=True)
produccion_rusia_gas = produccion_rusia_gas.drop(['Notes'], axis=1)
produccion_rusia_gas.tail()
```

```python
## Refinación de gasolina tipo motor en Arabia
refineryOutput_motor_arabia = quandl.get('JODI/OIL_GSROKB_SAU')
refineryOutput_motor_arabia.rename(columns={'Value': 'Refinación_Motor_ARABIA'}, inplace=True)
refineryOutput_motor_arabia = refineryOutput_motor_arabia.drop(['Notes'], axis=1)
refineryOutput_motor_arabia.tail()
```

```python
##Oferta Diesel

##Alemania
refineryOutput_diesel_alemania = quandl.get('JODI/OIL_GDROKB_DEU')
refineryOutput_diesel_alemania.rename(columns={'Value': 'Refinación_diesel_alemania'}, inplace=True)
refineryOutput_diesel_alemania = refineryOutput_diesel_alemania.drop(['Notes'], axis=1)

##USA
refineryOutput_diesel_usa = quandl.get('JODI/OIL_GDROKB_USA')
refineryOutput_diesel_usa.rename(columns={'Value': 'Refinación_diesel_USA'}, inplace=True)
refineryOutput_diesel_usa = refineryOutput_diesel_usa.drop(['Notes'], axis=1)

###Canadá
refineryOutput_diesel_canada = quandl.get('JODI/OIL_GDROKB_CAN')
refineryOutput_diesel_canada.rename(columns={'Value': 'Refinación_diesel_canada'}, inplace=True)
refineryOutput_diesel_canada= refineryOutput_diesel_canada.drop(['Notes'], axis=1)

###China
refineryOutput_diesel_china = quandl.get('JODI/OIL_GDROKB_CHN')
refineryOutput_diesel_china.rename(columns={'Value': 'Refinación_diesel_china'}, inplace=True)
refineryOutput_diesel_china= refineryOutput_diesel_china.drop(['Notes'], axis=1)


####Mergear los datos de producción y oferta de diesel de los principales países
mg1 = pd.merge(refineryOutput_diesel_alemania, refineryOutput_diesel_usa, on= 'Date')
mg2 = pd.merge(mg1, refineryOutput_diesel_china, on= 'Date')
oferta_diesel= pd.merge(mg2, refineryOutput_diesel_canada, on = 'Date')


##Adiciono columna con el total
oferta_diesel['Muestra_diesel'] = oferta_diesel['Refinación_diesel_alemania'] + oferta_diesel['Refinación_diesel_USA'] + oferta_diesel['Refinación_diesel_china'] + oferta_diesel['Refinación_diesel_canada']
oferta_diesel['Muestra_diesel'] = oferta_diesel['Muestra_diesel'].astype(int)

###Muestro el dataser completo
oferta_diesel.tail()
```

```python
##Llamando y renombrando los datos de niveles de producción de crudo


##Rusia
produccion_rusia_crudo = quandl.get('JODI/OIL_CRPRKB_RUS')
produccion_rusia_crudo.rename(columns={'Value': 'Produccion_crudo_RUSIA'}, inplace=True)
produccion_rusia_crudo = produccion_rusia_crudo.drop(['Notes'], axis=1)

##Arabia Saudita
produccion_arabia_crudo = quandl.get('JODI/OIL_CRPRKB_SAU')
produccion_arabia_crudo.rename(columns={'Value': 'Produccion_crudo_ARABIA'}, inplace=True)
produccion_arabia_crudo = produccion_arabia_crudo.drop(['Notes'], axis=1)

##China
produccion_china_crudo = quandl.get('JODI/OIL_CRPRKB_CHN')
produccion_china_crudo.rename(columns={'Value': 'Produccion_crudo_CHINA'}, inplace=True)
produccion_china_crudo = produccion_china_crudo.drop(['Notes'], axis=1)

##Canadá
produccion_canada_crudo = quandl.get('JODI/OIL_CRPRKB_CAN')
produccion_canada_crudo.rename(columns={'Value': 'Produccion_crudo_CANADA'}, inplace=True)
produccion_canada_crudo = produccion_canada_crudo.drop(['Notes'], axis=1)

##USA
produccion_usa_crudo = quandl.get('JODI/OIL_CRPRKB_USA')
produccion_usa_crudo.rename(columns={'Value': 'Produccion_crudo_USA'}, inplace=True)
produccion_usa_crudo = produccion_usa_crudo.drop(['Notes'], axis=1)


####Mergear los datos de producción y oferta de crudo de los principales países
mg1 = pd.merge(produccion_arabia_crudo, produccion_canada_crudo, on= 'Date')
mg2 = pd.merge(mg1, produccion_china_crudo, on = 'Date')
mg3= pd.merge(mg2, produccion_rusia_crudo, on= 'Date')
oferta_crudo = pd.merge(mg3, produccion_usa_crudo, on = 'Date')




### Adicionar una columna con la suma del total de los datos

oferta_crudo['Muestra_crudo'] = oferta_crudo['Produccion_crudo_ARABIA'] + oferta_crudo['Produccion_crudo_CANADA'] + oferta_crudo['Produccion_crudo_CHINA'] + oferta_crudo['Produccion_crudo_RUSIA'] + oferta_crudo['Produccion_crudo_USA']
oferta_crudo['Muestra_crudo'] = oferta_crudo['Muestra_crudo'].astype(int)

##Mostrar el conjunto de datos
oferta_crudo.tail()

```

```python
### Construcción de dataser completo

mg1= pd.merge(oferta_crudo, oferta_diesel, on= 'Date')
mg2= pd.merge(mg1, refineryOutput_fuelOil_canada, on= 'Date')
mg3= pd.merge(mg2, produccion_rusia_gas, on= 'Date')
niveles_pdn= pd.merge(mg3, refineryOutput_motor_arabia, on= 'Date')
niveles_pdn.tail()
```

```python
### Exportar a Excel los datos de producción

##Exporto lo construido a Excel para facilitar ensayos
outputFile = "Data/produccion.xlsx" ##Imprimiendo el conjunto de datos
niveles_pdn.to_excel(outputFile) # Only argument is a string of the output file path
```
